```python
zf=zipfile.ZipFile('software_private_income.zip')
```

```python
text_files=zf.infolist()
list_=[]
for text_file in text_files:
    print(text_file.filename)
    df=pd.read_csv(zf.open(text_file.filename), header=None, parse_dates=True)
    list_.append(df)

frame=pd.concat(list_, axis=0, ignore_index=True)
    
    
```

```python
frame
```

```python
frame[0].unique()
frame.dtypes
```

```python
frame_reshape=frame[[0,1,2,3]]
for i in range(4,(len(frame.columns)-1),2):
    frame_append=frame[[0,1,i,i+1]]
    frame_append.columns=frame_reshape.columns
    frame_reshape=frame_reshape.append(frame_append)
frame_reshape.columns
```

```python
frame_reshape.shape
```

```python
frame_reshape.columns=['company','description','date','value']
```

```python
frame_reshape.shape
frame_reshape['value']=pd.to_numeric(frame_reshape['value'], downcast='float',errors='coerce')
frame_reshape['date']=pd.to_datetime(frame_reshape['date'], errors='coerce')
```

```python
frame_reshape.dtypes
```

```python
frame_reshape
```

```python
frame_reshape['year']=frame_reshape['date'].dt.year
frame_reshape['month']=frame_reshape['date'].dt.month
```

```python
frame_reshape['year'].unique()
```

```python
frame_reshape['quarter']=np.where(frame_reshape['month']<4, 'Q1', 0)
frame_reshape['quarter']=np.where((frame_reshape['month']<7 )& (frame_reshape['month']>3), 'Q2', frame_reshape['quarter'])
frame_reshape['quarter']=np.where((frame_reshape['month']< 10 ) & (frame_reshape['month']>6),'Q3', frame_reshape['quarter'])
frame_reshape['quarter']=np.where(frame_reshape['month']>9, 'Q4', frame_reshape['quarter'])
```
