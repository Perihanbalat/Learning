```python
#merge all datasets available after being formated


directory_Path=('data_interim')
text_files=os.listdir('data_interim')
list_=[]
for text_file in text_files:
    print(text_file)
    df=pd.read_csv(text_file,  parse_dates=True)
    df["origin"]=text_file
    list_.append(df)

frame=pd.concat(list_, axis=0, ignore_index=True)
```

```python
frame
```

```python
frame.origin.unique()
```

```python
#Create a data frame for counts of non missing values for all features in all datasets available

no_na_data_counts=frame.groupby(['year','quarter','origin'], sort=True).count().reset_index()
no_na_data_counts.to_csv('no_navalues_counts.csv')

no_na_data_counts.head()
```

```python
no_na_data_ratio=no_na_data_counts.iloc[:,3:].div(no_na_data_counts.company, axis=0)
no_na_data_ratio['origin']=no_na_data_counts['origin'].to_numpy()
no_na_data_ratio.to_csv('no_navalues_ratio.csv')
no_na_data_ratio
```

```python
number_companies=frame.groupby(['year','origin'], sort=True).company.nunique()
```
